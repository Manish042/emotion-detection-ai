```python
'''mage_paths → list of all image file paths

labels → list of corresponding class labels

os.listdir(dir) → read all folders (labels)

os.path.isdir → check folder

Loop nested → folder → images

print(label, "completed") → show progress '''


def createDataFrame(dir):            
    image_paths = []
    labels = []

    for label in os.listdir(dir):
        label_path = os.path.join(dir, label)
        if os.path.isdir(label_path):
            for imagename in os.listdir(label_path):
                image_paths.append(os.path.join(label_path, imagename))
                labels.append(label)
        print(label, "completed")  # Ek label ke sare images ke baad print

    return image_paths, labels
```

```python
from tensorflow.keras.preprocessing.image import load_img    # type: ignore
import numpy as np 
from tqdm.notebook import tqdm

def extract_features(image_paths):
    features = []
    
    for img_path in tqdm(image_paths):
        # Correct: use color_mode instead of grayscale
        img = load_img(img_path, color_mode='grayscale')  
        img = np.array(img)
        features.append(img)
    
    # Convert to numpy array after loop
    features = np.array(features)
    
    # Reshape for CNN input (assuming 48x48 grayscale images)
    features = features.reshape(len(features), 48, 48, 1)
    
    return features
```

```python
train_features = extract_features(train['image'])   # Convert all training images to grayscale arrays and reshape them for CNN input
```

```python
test_features = extract_features(test['image'])
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical  # type: ignore
```

```python
le= LabelEncoder()    #Ye text labels ko numbers me convert karta hai
le.fit(train['label'])  
```

```python
y_train = le.transform(train['label'])
y_test = le.transform(test['label'])
```

```python
y_train = to_categorical(y_train,num_classes =7)
y_test = to_categorical(y_test,num_classes =7)
```

```python
from tensorflow.keras.models import Sequential # type: ignore
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout #type: ignore
```

```python
model = Sequential()
# 1️⃣ First Convolution Layer
model.add(Conv2D(128,kernel_size=(3,3), activation='relu', input_shape=(48,48,1)))  # conv2D -edge and shape detect 
model.add(MaxPooling2D(pool_size=(2,2)))     #MaxPooling2D-- reduce image 2*2
model.add(Dropout(0.4))                     #Dropout(0.4) → overfitting kam karta hai (40% neurons ignore)

# 2️⃣ Second Convolution Layer
model.add(Conv2D(256,kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))


# 3️⃣ Third Convolution Layer
model.add(Conv2D(512,kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

# 4️⃣ Flatten Layer
model.add(Flatten())   #2D feature map ko 1D vector me convert karta ha taki fully connected layer me ja sake 

# 5️⃣ Fully Connected Layer
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.3))

# 6️⃣ Output Layer
model.add(Dense(7,activation='softmax'))
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(
    x=x_train,
    y=y_train,
    batch_size=32,
    epochs=50,
    validation_data=(x_test, y_test)
)
```
